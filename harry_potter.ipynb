```python
import random      #Importar random
```

```python
contador1 = 0                #Contador para condiciones de victoria o derrrota
contador2 = 0                 
```

```python
# Creamos los hechizos para el random
a=["expelliarmus",                      
    "petrificus totalus",
    "imperio",
    "reducto",
    "crucio",
    "avada Kedavra",
    "sectumsempra",]
```

```python
d=[ "repello",
    "expecto patronum",
    "expelliarmus",
    "stupefy",
    "inpedimenta",
    "incarcerous"]
```

```python
jugador1 = input("Escribe tu nombre:").upper()           #Preguntamos a los jugadores sus nombres
jugador2 = input("Escribe tu nombre:").upper()
```

```python
turno = "jugador1"                  #Asignamos el turno a un jugador
while True:                          #Bucle de ejecucion del juego
    if turno == "jugador1":              #Condicion para que cambie de turno
        print("-🪄Turno actual", jugador1)
        hechizo = input(f" 🧙‍♀️ {jugador1} Quieres atacar(A), defender(D) o pulsa (S) para salir").lower() 
        if hechizo == "s":
            print("Has salido del juego.. \nHasta luego ¡¡MUGGLE!! 🔥 \n --------------------------- \n")                       #Condicion para salir del juego jugador1
            break
        if hechizo not in ["a", "d"]:                                                 #Condicion si no selecciona una opcion valida  jugador1
            print(f"Opcion no valida ⛔️ Elige (A) para atacar, (D) para defender o (S) para salir")     
            continue
        if hechizo ==  "a":                           #Condicion de ataque jugador1
            hechizo = random.choice(a)
            valores = random.randint(1,50)
            contador2 -= valores
            print(f"Has elegido atacar 🪄 tu hechizo es: {hechizo}, 🪫Daño: {valores}, {jugador2} te han quitado: {valores} vidas \n {jugador1} tu vida quedaria:{contador1} \n {jugador2} tu vida quedaria:{contador2}")
        else:                                          #Condicion para defenderse  jugador1
            hechizo = random.choice(d)
            valores = random.randint(1,50)
            contador1 += valores
            print(f"Has eligido defensa 🧙‍♀️ tu hechizo es: {hechizo}, ❤️Vida: {valores}, {jugador1} se te suman: {valores} vidas \n {jugador1} tu vida quedaria:{contador1} \n {jugador2} tu vida quedaria:{contador2}")
        turno = "jugador2"                            #Cambia el turno al siguiente jugador
    elif turno == "jugador2": 
        print("-Turno actual", jugador2)                   
        hechizo = input(f"🧙‍♀️ {jugador2} Quieres atacar(A), defender(D) o pulsa (S) para salir").lower()
        if hechizo == "s":                                                      #Condicion para salir del juego jugador2
            print("Has salido del juego.. \nHasta luego ¡¡MUGGLE!! 🫥")
            break 
        if hechizo not in ["a", "d"]:                                          #Condicion opcion no valida jugador2
            print(f"Opcion no valida ⛔️ Elige (A) para atacar, (D) para defender o (S) para salir")
            continue 
        if hechizo ==  "a":                           #Condicion de ataque  jugador2
            hechizo = random.choice(a)
            valores = random.randint(1,50)
            contador1 -= valores
            print(f"Has elegido atacar 🔥 tu hechizo es: {hechizo}, 🪫Daño: {valores}, {jugador1} te han quitado: {valores} vidas \n {jugador1} tu vida quedaria:{contador1} \n {jugador2} tu vida quedaria:{contador2}")
        else:                                            #Condicion de defensa jugador2
            hechizo = random.choice(d)
            valores = random.randint(1,50)
            contador2 += valores
            print(f"Has eligido defensa 🪄 tu hechizo es: {hechizo}, ❤️Vida: {valores}, {jugador2} se te suman: {valores} vidas \n {jugador1} tu vida quedaria:{contador1} \n {jugador2} tu vida quedaria:{contador2}")
        turno = "jugador1"                                      #Cambia el turno
    if contador1 >= 50 or contador2 <= -50:
        print("\n-------------------------------------\n")
        print(f"GANA !!{jugador1}¡¡ 🏆 \n")                   #Condicion para que gane el jugador 1
        print(f"{jugador2} 💀 HAS PERDIDO, SANGRE SUCIA ")
        break
    elif contador2 >= 50 or contador1 <= -50:
        print("\n-------------------------------------\n")
        print(f"GANA !!{jugador2}¡¡ 🏆 \n")                   #Condicion para que gane el jugador 1
        print(f"{jugador1} 💀 HAS PERDIDO, SANGRE SUCIA ")                   #Condicion para que gane el jugador 2
        break
```
